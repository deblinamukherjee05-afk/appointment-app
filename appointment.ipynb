### Appointment Data Model Definition

To effectively manage appointments, we will store the following key information for each appointment:

1.  **Date**: The specific date of the appointment. This will be stored as a `datetime.date` object or a string in 'YYYY-MM-DD' format for simplicity in initial representation.
2.  **Time**: The scheduled time of the appointment. This will be stored as a `datetime.time` object or a string in 'HH:MM' format.
3.  **Client Name**: The full name of the client. This will be a `string`.
4.  **Service**: The type of service booked (e.g., 'Haircut', 'Massage'). This will be a `string`.
5.  **Duration**: The length of the appointment in minutes. This will be an `integer`.
6.  **Status**: The current state of the appointment (e.g., 'Scheduled', 'Completed', 'Cancelled'). This will be a `string`.
7.  **Notes**: Any additional information or specific requests for the appointment. This will be a `string`.

Each individual appointment will be represented as a Python `dictionary`, where keys are the attribute names and values are the corresponding data. A collection of all appointments will be stored in a `list` of these dictionaries.

## Connect 'View Appointments' UI to backend

## Connect 'Update Appointment' UI to backend

### Subtask:
Define the callback function for the 'Update Appointment' button to retrieve input values and call the `update_appointment` function.

#### Instructions
1.  Define a callback function for the `update_button`. Inside this function:
    *   Clear the `output_area`.
    *   Retrieve the `index` from `update_index.value`. Ensure it's treated as an integer.
    *   Create a dictionary `kwargs` to hold the fields to update. Only include fields where the corresponding widget has a non-empty or non-default value.
        *   For `update_date`, `update_time`, `update_client_name`, `update_service`, `update_status`, `update_notes`, check if `value` is not an empty string.
        *   For `update_duration`, check if `value` is not 0 (or its initial default value).
    *   Call the `update_appointment()` function with the retrieved `index` and the `kwargs` dictionary using `**kwargs`.
    *   Clear the input fields of the update widgets after updating (set them back to empty or default values).
2.  Attach this callback function to the `update_button`'s `on_click` event.

**Reasoning**:
I need to implement the callback function for the 'Update Appointment' button, which will gather input from the update widgets, construct a dictionary of fields to update, and then call the `update_appointment` backend function.



```python
def on_update_button_clicked(button):
    with output_area:
        clear_output()
        try:
            index = update_index.value
            kwargs = {}

            if update_date.value: kwargs["Date"] = update_date.value
            if update_time.value: kwargs["Time"] = update_time.value
            if update_client_name.value: kwargs["Client Name"] = update_client_name.value
            if update_service.value: kwargs["Service"] = update_service.value
            # Check if duration is explicitly set to a non-zero value, or if it's 0 and not the default (optional logic)
            if update_duration.value != 0: kwargs["Duration"] = update_duration.value # Assuming 0 is a default/empty state for update
            if update_status.value: kwargs["Status"] = update_status.value
            if update_notes.value: kwargs["Notes"] = update_notes.value

            if not kwargs:
                print("No update fields provided. Please enter values to update.")
                return

            update_appointment(index, **kwargs)

            # Clear input fields after updating
            update_index.value = 0 # Reset to default
            update_date.value = ''
            update_time.value = ''
            update_client_name.value = ''
            update_service.value = ''
            update_duration.value = 0 # Reset to default
            update_status.value = ''
            update_notes.value = ''

        except Exception as e:
            print(f"Error updating appointment: {e}")

# Attach the callback function to the update_button's on_click event
update_button.on_click(on_update_button_clicked)

print("Update Appointment button callback defined and attached.")
```
